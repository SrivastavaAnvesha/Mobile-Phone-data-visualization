### Creating Visualizations with Matplotlib

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Basic Level

```python
# Import data Mobilephone_data_final_V2 as df.
df=pd.read_csv(r'Mobilephone_data_final_V2.csv')
df
```

```python
# How many distinct operating systems are present in the data.
df['Operating system'].nunique()
```

```python
# Print the value counts of Brand.
df['Brand'].value_counts()
```

```python
# Print the MAX price of all mobiles.
df['Brand'].value_counts()
```

#### Intermediate Level

```python
# Make a scatter plot chart to show the relationship between rear and front camera resolutions.
plt.figure(figsize=(5, 3))
plt.scatter(df['Resolution x'], df['Resolution y'], color='blue', alpha=0.5)
plt.title('Relationship Between Rear and Front Camera Resolutions')
plt.xlabel('Rear Camera Resolution (MP)')
plt.ylabel('Front Camera Resolution (MP)')
plt.show()
```

```python
# Print the average price of each Brand.
df.groupby('Brand')['Price'].mean()
```

```python
# Print the Operating system wise each Brand Max price.
df.groupby(['Operating system', 'Brand'])['Price'].max()
```

```python
# Create a histogram to see the distribution of phone prices.
plt.figure(figsize=(4, 3))
plt.hist(df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Phone Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Display the distribution of battery capacities across different brands using boxplot.
plt.figure(figsize=(10, 3))
sns.boxplot(x='Brand', y='Battery capacity (mAh)', data=df)
plt.title('Distribution of Battery Capacities Across Brands')
plt.xlabel('Brand')
plt.ylabel('Battery Capacity (mAh)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Explore the relationship between screen size and price.
plt.figure(figsize=(9, 3))
plt.scatter(df['Screen size (inches)'], df['Price'], color='purple', alpha=0.5)
plt.title('Relationship Between Screen Size and Price')
plt.xlabel('Screen Size (inches)')
plt.ylabel('Price')
plt.show()
```

#### Advanced Level

```python
# Make a bar chart to compare the average RAM across different brands. 
average_ram = df.groupby('Brand')['RAM (MB)'].mean()
plt.figure(figsize=(10, 3))
average_ram.plot(kind='bar', color='skyblue')
plt.title('Average RAM Across Different Brands')
plt.xlabel('Brand')
plt.ylabel('Average RAM (GB)')
plt.xticks(rotation=45)
plt.show()
```
